```python
import pandas as pd

data_cr = {
    'x1': [0, 0, 0, 0],
    'x2': [1, 0, 0, 0],
    'x3': [0, 1, 0, 0],
    'x4': [0, 1, 0, 0]
}
index_cr = ['x1', 'x2', 'x3', 'x4']
df_cr = pd.DataFrame(data_cr, index=index_cr)
print('CR:')
print(df_cr)

data_ca = {
    'x1': [0, 0, 0, 0],
    'x2': [0, 0, 0, 0],
    'x3': [0, 1, 0, 0],
    'x4': [0, 0, 0, 0]
}
index_ca = ['x1', 'x2', 'x3', 'x4']
df_ca = pd.DataFrame(data_ca, index=index_ca)
print('CA:')
print(df_ca)

def calculate_stc(df_cr, df_ca):
    satisfied_requirements = ((df_cr > 0) & (df_ca > 0)).sum().sum()
    total_requirements = (df_cr > 0).sum().sum()
    return satisfied_requirements / total_requirements if total_requirements > 0 else 0

stc_value = calculate_stc(df_cr, df_ca)

print(f"STC Value: {stc_value}")
```

```python
import pandas as pd

mc_cr_data = {
    'x1': [0, 0, 0, 0],
    'x2': [0, 0, 1, 1],
    'x3': [0, 0, 0, 0],
    'x4': [0, 0, 0, 0]
}

mc_ca_data = {
    'x1': [0, 0, 0, 0],
    'x2': [0, 0, 1, 0],
    'x3': [0, 0, 0, 0],
    'x4': [0, 0, 0, 0]
}

mc_cr_df = pd.DataFrame(mc_cr_data, index=['x1', 'x2', 'x3', 'x4'])
mc_ca_df = pd.DataFrame(mc_ca_data, index=['x1', 'x2', 'x3', 'x4'])

security_devs = ['x3', 'x4']
devs = ['x1', 'x2'] 

def mc_diff(mc_cr_df, mc_ca_df, devs, security_devs):
    diff_count = 0
    for dev in devs:
        for sec_dev in security_devs:
            if mc_cr_df.loc[dev, sec_dev] > 0 and mc_ca_df.loc[dev, sec_dev] > 0:
                diff_count += 1
            if mc_cr_df.loc[sec_dev, dev] > 0 and mc_ca_df.loc[sec_dev, dev] > 0:
                diff_count += 1
    return diff_count

def mc_cr_total(mc_cr_df, devs, security_devs):
    total_count = 0
    for dev in devs:
        for sec_dev in security_devs:
            if mc_cr_df.loc[dev, sec_dev] > 0:
                total_count += 1
            if mc_cr_df.loc[sec_dev, dev] > 0:
                total_count += 1
    return total_count

def mc_stc(mc_cr_df, mc_ca_df, devs, security_devs):
    diff = mc_diff(mc_cr_df, mc_ca_df, devs, security_devs)
    total = mc_cr_total(mc_cr_df, devs, security_devs)
    if total == 0:
        return 0
    return diff / total

print('2C-CR:')
print(mc_cr_df)
print('2C-CA:')
print(mc_ca_df)
print('security devs:')
print(security_devs)
mc_stc_value = mc_stc(mc_cr_df, mc_ca_df, devs, security_devs)
print(f"MC-STC: {mc_stc_value}")
```
